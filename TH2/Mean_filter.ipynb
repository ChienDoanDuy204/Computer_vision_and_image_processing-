```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
FILE_PATH = '/content/drive/MyDrive/XLA&TGMT/data/B5EhD.png'
```

```python
img = cv2.imread(FILE_PATH,flags=0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
img_result = np.ones(img.shape)
```

```python
def mean_filter(img,img_result,thress_hold):
  for i in range(img.shape[0]-2):
    for j in range(img.shape[1]-2):
      img_result[i+1,j+1] = np.mean(img[i:i+3,j:j+3]) if np.abs(img[i+1,j+1] - np.mean(img[i:i+3,j:j+3])) >thress_hold else img[i+1,j+1]
  return img_result
```

```python
result = mean_filter(img,img_result,50)
```

```python
plt.imshow(result,cmap='gray')
```
